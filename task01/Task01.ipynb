```python
import pandas as pd
df = pd.read_excel("Product-Sales-Region.xlsx")
df["Promotion"] = df["Promotion"].fillna("NoPromotion")
```

```python
for col in ["Quantity", "UnitPrice", "Discount", "TotalPrice", "ShippingCost"]:
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)
```

```python
df["DeliveryDays"] = (df["DeliveryDate"] - df["OrderDate"]).dt.days
df["DiscountAmount"] = df["Quantity"] * df["UnitPrice"] * df["Discount"]
df["RevenuePerUnit"] = df["TotalPrice"] / df["Quantity"]
df["IsPromotionUsed"] = (df["Promotion"] != "NoPromotion").astype(int)
df["OrderMonth"] = df["OrderDate"].dt.month_name()
df.to_excel("Product-Sales-Region-Cleaned.xlsx", index=False)
print("Done:", df.shape)
```
